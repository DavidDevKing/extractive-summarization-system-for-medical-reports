```python
!git clone "https://github.com/DavidDevKing/extractive-summarization-system-for-medical-reports.git"
!pip install -r extractive-summarization-system-for-medical-reports/notebooks/requirements.txt

import sys
sys.path.append("extractive-summarization-system-for-medical-reports/notebooks/")
from utils import *


from datasets import load_dataset
from rouge_score import rouge_scorer


import re
from tqdm.auto import tqdm






# Load 50 samples from the TEST split
print("Loading 50 test records...")
dataset = load_dataset("ccdv/pubmed-summarization", "section", split="test", streaming=True)
test_records = list(dataset.take(50))

# Evaluation Loop
all_r1 = []
all_r2 = []
all_rL = []

print("Evaluating Baseline (Unsupervised K-Means)...")

for entry in tqdm(test_records):
    article_text = entry['article']
    reference_abstract = entry['abstract']

    # Generate summary
    candidate_summary = extract_summary(article_text, num_sentences=3)

    # Get the string scores from your utils function
    score_string = extract_scores(reference_abstract, candidate_summary)

    # Use Regex to find the F1-Scores for rouge1 and rougeL
    # Looks for the number immediately following 'rouge1: F1-Score = '
    r1_match = re.search(r"rouge1: F1-Score = ([\d\.]+)", score_string)
    r2_match = re.search(r"rouge1: F1-Score = ([\d\.]+)", score_string)
    rL_match = re.search(r"rougeL: F1-Score = ([\d\.]+)", score_string)

    if r1_match and r2_match and rL_match:
        all_r1.append(float(r1_match.group(1)))
        all_r2.append(float(r2_match.group(1)))
        all_rL.append(float(rL_match.group(1)))

# Final Results
if len(all_r1) > 0:
    avg_r1 = sum(all_r1) / len(all_r1)
    avg_r2 = sum(all_r2) / len(all_r2)
    avg_rL = sum(all_rL) / len(all_rL)

    print("\n" + "="*50)
    print("--- FINAL BASELINE AVERAGES (50 SAMPLES) ---")
    print(f"Average ROUGE-1 F1: {avg_r1:.4f}")
    print(f"Average ROUGE-2 F1: {avg_r2:.4f}")
    print(f"Average ROUGE-L F1: {avg_rL:.4f}")
    print("="*50)
else:
    print("No scores were collected. Check your regex or utility function.")
```
